## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
cur.execute("SELECT DISTINCT Launch_Site FROM SPACEXTBL;")
result = cur.fetchall()

result
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
sql_query = """
SELECT * FROM SPACEXTBL
WHERE Launch_Site
LIKE 'CCA%' limit 5;
"""

# Execute the SQL query
cur.execute(sql_query)

# Fetch the result if needed
result = cur.fetchall()
result
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
sql_query = """
SELECT Customer, SUM(PAYLOAD_MASS__KG_)
FROM SPACEXTBL
WHERE Customer = 'NASA (CRS)';
"""

# Execute the SQL query
cur.execute(sql_query)

# Fetch the result if needed
result = cur.fetchall()
result
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
sql_query = """
SELECT Booster_Version, AVG(PAYLOAD_MASS__KG_) AS avg_payload_mass
FROM SPACEXTBL
WHERE Booster_Version = 'F9 v1.1';
"""

# Execute the SQL query
cur.execute(sql_query)

# Fetch the result if needed
result = cur.fetchall()
result
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
sql_query = """
SELECT MIN(Date)
FROM SPACEXTBL
WHERE Landing_Outcome = 'Success (ground pad)';

"""

# Execute the SQL query
cur.execute(sql_query)

# Fetch the result if needed
result = cur.fetchall()
result
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
sql_query = """
SELECT Booster_Version
FROM SPACEXTBL
WHERE Landing_Outcome = 'Success (drone ship)'
AND PAYLOAD_MASS__KG_ > 4000
AND PAYLOAD_MASS__KG_ < 6000;

"""

# Execute the SQL query
cur.execute(sql_query)

# Fetch the result if needed
result = cur.fetchall()
result
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
sql_query = """
SELECT Mission_Outcome, COUNT(*) AS Total
FROM SPACEXTBL
WHERE Mission_Outcome IN ('Success', 'Success (drone ship)', 'Success (payload status unclear)', 'Failure (in flight)')
GROUP BY Mission_Outcome;

"""
# Execute the SQL query
cur.execute(sql_query)

# Fetch the result if needed
result = cur.fetchall()
result
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
sql_query = """
SELECT Booster_Version
FROM SPACEXTBL
WHERE PAYLOAD_MASS__KG_ = (
    SELECT MAX(PAYLOAD_MASS__KG_)
    FROM SPACEXTBL
);

"""
# Execute the SQL query
cur.execute(sql_query)

# Fetch the result if needed
result = cur.fetchall()
result
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
sql_query = """
SELECT 
    CASE
        WHEN substr(Date, 6, 2) = '01' THEN 'January'
        WHEN substr(Date, 6, 2) = '02' THEN 'February'
        WHEN substr(Date, 6, 2) = '03' THEN 'March'
        WHEN substr(Date, 6, 2) = '04' THEN 'April'
        WHEN substr(Date, 6, 2) = '05' THEN 'May'
        WHEN substr(Date, 6, 2) = '06' THEN 'June'
        WHEN substr(Date, 6, 2) = '07' THEN 'July'
        WHEN substr(Date, 6, 2) = '08' THEN 'August'
        WHEN substr(Date, 6, 2) = '09' THEN 'September'
        WHEN substr(Date, 6, 2) = '10' THEN 'October'
        WHEN substr(Date, 6, 2) = '11' THEN 'November'
        WHEN substr(Date, 6, 2) = '12' THEN 'December'
    END AS MonthName,
    Landing_Outcome AS FailureLandingOutcome,
    Booster_Version,
    Launch_Site
FROM SPACEXTBL
WHERE substr(Date, 1, 4) = '2015'
    AND Landing_Outcome LIKE '%Failure (drone ship)%';


"""
# Execute the SQL query
cur.execute(sql_query)

# Fetch the result if needed
result = cur.fetchall()
result
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
sql_query = """
SELECT Landing_Outcome, COUNT(*) AS OutcomeCount
FROM SPACEXTBL
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY Landing_Outcome
ORDER BY OutcomeCount DESC;
"""
# Execute the SQL query
cur.execute(sql_query)

# Fetch the result if needed
result = cur.fetchall()
result
```
